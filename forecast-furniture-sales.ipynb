## D Check Stationary statistically

```python
ad_fuller_result_1 = adfuller(df['Sales'].diff()[1:])

print('Sales')
print(f'ADF Statistic: {ad_fuller_result_1[0]}')
print(f'p-value: {ad_fuller_result_1[1]}')

print('\n---------------------\n')

ad_fuller_result_2 = adfuller(df['Profit'].diff()[1:])

print('Profit')
print(f'ADF Statistic: {ad_fuller_result_2[0]}')
print(f'p-value: {ad_fuller_result_2[1]}')

print('\n---------------------\n')

ad_fuller_result_3 = adfuller(df['Discount'].diff()[1:])

print('Discount')
print(f'ADF Statistic: {ad_fuller_result_2[0]}')
print(f'p-value: {ad_fuller_result_2[1]}')

print('\n---------------------\n')

ad_fuller_result_4 = adfuller(df['Quantity'].diff()[1:])

print('Quantity')
print(f'ADF Statistic: {ad_fuller_result_2[0]}')
print(f'p-value: {ad_fuller_result_2[1]}')
```

### If p-value is less than 0.05 that means data is stationary, and vice versa¶
### So, It's clear that data is not stationary



## E Check Causality

```python
#For example between Profit and Discount
print('Discount causes Profit?\n')
print('------------------')
granger_1 = grangercausalitytests(df[['Discount', 'Profit']], 4)

print('\Profit causes Discount?\n')
print('------------------')
granger_2 = grangercausalitytests(df[['Profit', 'Discount']], 4)
```

### 3- Many stats methods can be used. ex: to get p-value to make sure if data stationary or not

### 4- If p-value >= 0.05, Data is not stationary
